```python
import numpy as np
```

```python
x = np.load("../csv_data/tri_an_thanh_hoa/x_direct.npy")
y = np.load("../csv_data/tri_an_thanh_hoa/y_direct.npy")
```

```python
print(x.shape)
print(y.shape)
```

```python
def check_sum_ne1(x):
    total = x.size
    print("tong số phần tử : ",total)
    num_ne1 = np.sum(x == -1)
    num_ne9999 = np.sum(x == -9999)

    print(f"Số phần tử = -1: {num_ne1}")
    print(f"Số phần tử = -9999: {num_ne9999}")
```

```python
check_sum_ne1(x)
```

```python
check_sum_ne1(y)
```

```python
import numpy as np
import matplotlib.pyplot as plt

start = 800
n_show = 10

for t in range(start, start + n_show):
    map_t = y[t]   # shape (34, 42)

    map_t = map_t.astype(float)
    map_t[map_t == -9999] = np.nan

    plt.figure(figsize=(6, 5))
    plt.imshow(map_t, cmap="viridis")
    plt.colorbar(label="Rain value")
    plt.title(f"y[{t}]")
    plt.axis("off")
    plt.show()
```
